   
## Turismo rural por provincias - Nov. 2021

```python
provincias = [pol['properties']['description'] for pol in data['features'] ]
turismo = [5376, 2326, 920, 2458, 9767, 9772, 2778, 4917, 11190, 1546, 7082, 12027, 1593, 9722, 3177, 2977, 3060, 1304, 3264, 3921, 12982, 1793, 5379, 2789, 4003, 2131, 4648, 5913, 1852, 10414, 4842, 2846, 7456, 1930, 2127, 703, 1759, 2254, 4607, 2151, 7829, 1583, 5315, 2472, 5070, 5106, 3857, 2975, 2528, 2863, 1202, 903]
prov_turismo = dict(zip(provincias,turismo))


for location_info in data['features']:
    location_info['properties']['turismo'] = prov_turismo[location_info['properties']['description']]
```

```python
data
```

```python
import branca.colormap as cmp

colores = cmp.LinearColormap(
    ['lightcyan','lightblue', 'steelblue', 'blue'],
#     vmin = round(min(turismo), -1), vmax = round(max(turismo), -3),
    vmin = min(turismo), vmax = max(turismo),
    caption = 'Turismo rural en Espa√±a'
)

def estilo_provincias (feature):

    return{ 'radius': 7,
        'fillColor': colores(feature['properties']['turismo']), 
        'color': colores(feature['properties']['turismo']), 
        'weight': 5,
        'opacity' : 1,
        'fillOpacity' : 0.8}
```

```python
m = folium.Map(location=[40.4309, -3.6878], zoom_start = 6)

folium.GeoJson(data, style_function = estilo_provincias,
               tooltip = folium.GeoJsonTooltip(fields = ['description', 'turismo'],
                                                    aliases=['Provincia: ', 'Turistas: '],
                                                    style = ("background-color: white; color: #333333; font-family: arial; font-size: 12px; padding: 10px;"),
                                                    sticky = True
                    )).add_to(m)

m
```
