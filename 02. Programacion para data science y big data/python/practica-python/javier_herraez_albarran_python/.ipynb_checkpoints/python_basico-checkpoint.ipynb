### 1. Escribe un programa que calcule todos los enteros entre 4000 y 9600 (ambos incluidos) que sean divisibles entre 5 pero no múltiplos de 7.

```python
num_1 = 4000
num_2 = 9600
[number for number in range(num_1, num_2 + 1) if (number % 5 == 0 and number % 7 != 0)]
#obtenemos los números a partir de un bucle que recorre desde num_1 hasta num_2 (se usa '+1' para que se incluya)
#a través de un if con las dos condiciones verificamos que cumplan las condiciones
```

### 2. Escribe una función llamada “enteros_absolutos” que reciba como entrada una lista de números enteros y genere a la salida otra lista con el valor absoluto de cada elemento distinto de 0 de la lista de entrada. Los 0 de la lista original deberán eliminarse en la salida.  

```python
def enteros_absolutos(lista):
    return [abs(number) for number in lista if number != 0]

lista_prueba = list(range(-5,5))
print(lista_prueba)
enteros_absolutos(lista_prueba)
```

### 3. Escribe una función llamada “comprimir” que reciba como parámetro una lista de elementos y devuelva como salida una lista de tuplas de la forma (elemento, número de repeticiones consecutivas).

```python
def comprimir(lista):
    tuplas = []
    contador = 1
    for number in lista:
        if not tuplas:
            tuplas.append((number, contador))
        elif number != tuplas[-1][0]:
            contador = 1
            tuplas.append((number, contador))
        else:
            tuplas.pop()
            contador += 1
            tuplas.append((number, contador))
    return tuplas

lista_prueba = [1, 1, 1, 2, 2, 3, 4, 4, 1, 1, 1]
#lista_prueba = ['a', 'a', 'a', 'b', 'b', 'c', 'd', 'd', 'a', 'a', 'a']
comprimir(lista_prueba)
```

### 4. Crea una función llamada “alumnos_aprobados” que reciba como parámetro un diccionario cuyas claves son nombres de alumnos y sus valores una lista de calificaciones numéricas. La salida de la función deberá ser una lista que contiene los nombres de los alumnos cuya nota media es igual o superior a 7 y NO tienen ningún suspenso.

```python
def alumnos_aprobados(alumnos):
    lista_aprobados = []
    for alumno in alumnos:
        if (sum(alumnos[alumno]) / len(alumnos[alumno]) >= 7) and (True not in [(nota < 5) for nota in alumnos[alumno]]):
            lista_aprobados.append(alumno)
    return lista_aprobados

dict_prueba = {'Juan': [8, 7, 6], 'Maria': [5, 10, 5], 'Pedro': [10, 3, 10]}
alumnos_aprobados(dict_prueba)
```

### 5. Escribe una función “habitantes” que reciba como entrada un nombre de ciudad y un diccionario cuyas claves son nombres de personas y su valor la ciudad en la que viven. La salida de la función deberá ser la lista de nombres de personas ordenados alfabéticamente del diccionario que viven en la ciudad recibida como. 

```python
def habitantes(ciudad, personas):
    lista_habitantes = [nombre for nombre in personas if personas[nombre] == ciudad]
    lista_habitantes.sort()
    return lista_habitantes

habitantes('Madrid', {'Juan': 'Madrid', 'Alicia': 'Madrid', 'Pedro': 'Cádiz'})
```
