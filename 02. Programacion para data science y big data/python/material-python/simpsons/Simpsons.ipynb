## Leemos todos los ficheros y nos aseguramos de que están bien leídos

```python
characters = pd.read_csv('characters.csv', sep=';')
episodes = pd.read_csv('episodes.csv', sep=';')
locations = pd.read_csv('locations.csv', sep=';')
scripts = pd.read_csv('scripts.csv', sep=';')
```

## ¿Cuál es la temporada con más episodios?

```python
episodes.groupby('season')['season'].size().sort_values(ascending=False).head()
```

## ¿Cual es la primera frase de la serie? ¿Quién la dice?

```python
temp_one = episodes[(episodes.season == 1) & (episodes.number_in_season == 1)]
scripts_episode_one = pd.merge(scripts, episode_one, left_on='episode_id', right_on='id', suffixes=('_script', '_episode'))
scripts_spoken_one = scripts_episode_one[scripts_episode_one.speaking_line == True]
scripts_spoken_one.sort_values('number').head(1)
```

## ¿Cuáles son los 10 personajes que más hablan en los episodios?

```python
character_lines = pd.merge(characters, scripts[scripts.speaking_line == True], left_on='id', right_on='character_id', suffixes=('_char', '_script'))
character_lines.groupby('name').size().sort_values(ascending=False).head(10)
```

## ¿Cuáles son las 10 localizaciones más recurrentes en los episodios?

```python
location_lines = pd.merge(locations, scripts, left_on='id', right_on='location_id', suffixes=('_location', '_script'))
location_lines.groupby('name').size().sort_values(ascending=False).head(10)
```

## ¿Cuál es el mejor episodio según IMDB?

```python
episodes.sort_values('imdb_rating', ascending = False).head()[['title','imdb_rating','imdb_votes']]
```

## ¿Cuál es la mejor temporada (en media) según IMDB?

```python
episodes.groupby('season')['imdb_rating'].mean().sort_values(ascending=False).head()
```
